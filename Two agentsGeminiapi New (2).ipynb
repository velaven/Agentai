```python
import os
from google import genai
from google.genai import types

def run_two_agents():
    # Initialize the client (it automatically picks up GEMINI_API_KEY from the environment)
    client = genai.Client()
    
    # We use the standard recommended model for general text tasks
    model_id = "gemini-2.5-flash"
    
    topic = "The impact of quantum computing on modern cybersecurity"
    
    print(f"🚀 Starting multi-agent workflow for topic: '{topic}'\n")

    # ----------------------------------------------------------------
    # AGENT 1: The Researcher
    # ----------------------------------------------------------------
    print("🤖 Agent 1 (Researcher) is gathering facts...")
    
    researcher_config = types.GenerateContentConfig(
        system_instruction=(
            "You are an expert technical researcher. Your job is to provide "
            "a bulleted list of raw, highly accurate facts, technical challenges, "
            "and recent breakthroughs regarding the user's topic. Do not write an essay."
        ),
        temperature=0.2, # Low temperature for factual accuracy
    )
    
    research_response = client.models.generate_content(
        model=model_id,
        contents=f"Research the following topic thoroughly: {topic}",
        config=researcher_config
    )
    
    raw_research_notes = research_response.text
    print("\n[Researcher Notes Completed]")
    
    # ----------------------------------------------------------------
    # AGENT 2: The Writer
    # ----------------------------------------------------------------
    print("\n🤖 Agent 2 (Writer) is drafting the final article...")
    
    writer_config = types.GenerateContentConfig(
        system_instruction=(
            "You are a professional technology journalist. Your job is to take raw "
            "research notes and turn them into an engaging, easy-to-understand, "
            "and polished newsletter article. Use headings and bullet points for readability."
        ),
        temperature=0.7, # Higher temperature for creative/engaging writing style
    )
    
    # We pass the output of Agent 1 directly into the prompt for Agent 2
    writer_prompt = f"""
    Here are the raw research notes you need to transform into an article:
    
    {raw_research_notes}
    """
    
    final_response = client.models.generate_content(
        model=model_id,
        contents=writer_prompt,
        config=writer_config
    )
    
    # ----------------------------------------------------------------
    # Output the Final Result
    # ----------------------------------------------------------------
    print("\n================= FINAL AGENT OUTPUT =================")
    print(final_response.text)
    print("======================================================")

if __name__ == "__main__":
    # Quick sanity check for the API key
    if not os.environ.get("GEMINI_API_KEY"):
        print("❌ Error: GEMINI_API_KEY environment variable not set.")
    else:
        run_two_agents()
```
